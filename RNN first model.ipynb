```
import numpy as np
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```
import keras
```

```
data1=[[i for i in range (0,100)]]
```

```
data1=np.array(data1,dtype=float)
```

```
type(data1)
```

```
target1=[[i for i in range(1,101)]]
target1=np.array(target1,dtype=float)
type(target1)
```

```
data1=data1.reshape(1,1,100)
target1=target1.reshape(1,1,100)
```

```
data1
```

```
test_x=[[i for i in range(100,200)]]
test_x=np.array(test_x,dtype=float)
type(test_x)
```

```
test_y=[[i for i in range(200,300)]]
test_y=np.array(test_y)
type(test_y)
```

```
model=Sequential()
```

```
test_x=test_x.reshape(1,1,100)
test_y=test_y.reshape(1,1,100)
```

```
model.add(LSTM(100,input_shape=(1,100),return_sequences=True))
model.add(Dense(100))
```

```
model.compile(loss='mean_absolute_error', optimizer='adam', metrics='accuracy')
```

```
model.fit(data1,target1,epochs=30,batch_size=1,verbose=2,validation_data=(test_x,test_y)) # Usually we should give 1000 epochs
```
